# SORU1

Senaryo: Bir restoranın yemek siparişi maliyeti hesaplama

Kurallar:

0 adet yemek için temel maliyet 50 TL (servis ücreti)

Her yemek maliyete 20 TL ekler.

Her içecek maliyete 5 TL ekler.

1. Soru: get_order_cost() fonksiyonunu yazın. Bu fonksiyonun iki parametresi olacak: meals (yemek sayısı) ve drinks (içecek sayısı). Bu fonksiyon toplam maliyeti döndürecek.

        * 2 adet yemek ve 0 adet içecek için toplam maliyet nedir?
   
        * 3 adet yemek ve 2 adet içecek için toplam maliyet nedir?
   
        * 4 adet yemek ve 1 adet içecek için toplam maliyet nedir?

```python
def get_order_cost(meals: int, drinks: int) -> float:
    """
    Calculate the total cost of a restaurant order.

    Parameters:
    meals (int): Number of meals to be included in the order.
    drinks (int): Number of drinks to be included in the order.

    Returns:
    float: Total cost of the order in Turkish Lira (TL).
    """

    # Define base cost for 0 meals (service fee)
    service_fee = 50

    # Calculate total meal cost by adding 20 TL per meal to the service fee
    total_meals_cost = service_fee + (meals * 20)

    # Calculate total drink cost by adding 5 TL per drink to the current total cost
    total_drinks_cost = total_meals_cost + (drinks * 5)

    # Return the final total order cost
    return total_drinks_cost

# Test the function with example inputs
print(get_order_cost(2, 0))  # Output: 70.0
print(get_order_cost(3, 2))  # Output: 100.0
print(get_order_cost(4, 1))  # Output: 90.0
```

# SORU2

Senaryo: Bir aracın kiralama maliyeti hesaplama

Kurallar:

0 gün kiralama için temel maliyet 100 TL (sabit ücret)

Her gün için 200 TL ekler.

Her kilometre başına 1 TL ekler.

1. Soru: get_rental_cost() fonksiyonunu yazın. Bu fonksiyonun iki parametresi olacak: days (gün sayısı) ve kilometers (kilometre sayısı). Bu 

fonksiyon toplam maliyeti döndürecek.

        * 3 gün ve 150 kilometre için toplam maliyet nedir?

        * 5 gün ve 300 kilometre için toplam maliyet nedir?

```python
def get_rental_cost(days: int, kilometers: int) -> float:
    """Calculates the rental cost of a car.

    Parameters:
        days (int): The number of days the car is rented.
        kilometers (int): The total distance driven in kilometers.

    Returns:
        float: The total cost of renting the car for the specified period and distance.

    Assumptions:
     - Base cost for 0 days rental is 100 TL
     - Additional daily cost is 200 TL
     - Cost per kilometer is 1 TL

    Examples:
    >>> get_rental_cost(3, 150)
    1450.0
    >>> get_rental_cost(5, 300)
    2700.0
    """
    
    # Base cost for 0 days rental is 100 TL
    base_cost = 100

    # Additional daily cost is 200 TL
    daily_cost = 200 * days

    # Cost per kilometer is 1 TL
    km_cost = kilometers * 1

    # Total cost calculation
    total_cost = base_cost + daily_cost + km_cost

    return total_cost


# Test the function with provided examples
print(get_rental_cost(3, 150))  # Expected output: 1450.0
print(get_rental_cost(5, 300))  # Expected output: 2700.0
```

# SORU3

Senaryo: Bir sinema bileti maliyeti hesaplama

Kurallar:

0 bilet için maliyet 0 TL

Her bilet için 50 TL ekler.

Öğrenciler için her bilette 10 TL indirim yapılır.

1. Soru: get_ticket_cost() fonksiyonunu yazın. Bu fonksiyonun iki parametresi olacak: tickets (bilet sayısı) ve students (öğrenci bilet 

sayısı). Bu fonksiyon toplam maliyeti döndürecek.

        * 1 bilet ve 1 öğrenci bileti için toplam maliyet nedir?

        * 3 bilet ve 1 öğrenci bileti için toplam maliyet nedir?

        * 5 bilet ve 3 öğrenci bileti için toplam maliyet nedir?

```python
def get_ticket_cost(tickets: int, students: int) -> float:
    """
    Calculates and returns the total cost of tickets based on number of adult tickets and student tickets.
    
    Parameters:
        tickets (int): The number of adult tickets.
        students (int): The number of student tickets.

    Returns:
        float: The total cost of all tickets after applying a discount for each student ticket.

    Raises:
        ValueError: If the number of tickets or students is negative.

    Example:
        >>> get_ticket_cost(1, 1)
        40.0
        >>> get_ticket_cost(3, 1)
        150.0
        >>> get_ticket_cost(5, 3)
        220.0
    """
    
    # Check if input is valid
    if tickets < 0 or students < 0:
        raise ValueError("Number of tickets and students must be non-negative.")
    
    # Calculate total cost for adult tickets
    total_cost = 50 * tickets

    # Apply discount for student tickets and add them into the total cost
    total_cost -= 10 * students

    return total_cost

print(get_ticket_cost(1,1)) 
print(get_ticket_cost(3,1)) 
print(get_ticket_cost(5,3)) 
```

# SORU4

Senaryo: Bir seyahat planının maliyeti hesaplama

Kurallar:

Temel maliyet: 500 TL (organize ücret).

Uçuş maliyeti: her uçuş için 1000 TL.

Konaklama maliyeti: her gece için 300 TL.

Tur rehberi maliyeti: günlük 200 TL.

1. Soru: calculate_trip_cost() fonksiyonunu yazın. Bu fonksiyon üç parametre alacak: flights (uçuş sayısı), nights (konaklama gecesi sayısı) 

ve guide_days (tur rehberi gün sayısı). Bu fonksiyon toplam maliyeti döndürecek.

        * 2 uçuş, 5 gece konaklama ve 3 gün tur rehberi için toplam maliyet nedir?

        * 3 uçuş, 4 gece konaklama ve 1 gün tur rehberi için toplam maliyet nedir?

```python
def calculate_trip_cost(flights: int, nights: int, guide_days: int) -> int:
    """
    Calculates the total cost of a trip based on number of flights,
    nights spent in accommodation, and days with tour guide service.
    
    Parameters:
        flights (int): The number of flights taken during the trip. 
            Must be a non-negative integer.
        
        nights (int): The number of nights spent at an accommodation. 
            Must be a non-negative integer.
            
        guide_days (int): The number of days with a tour guide service. 
            Must be a non-negative integer.

    Returns:
        int: The total cost of the trip, in Turkish Lira (TL).

    Raises:
        ValueError: If any input is less than zero.

    Examples:
        >>> calculate_trip_cost(2, 5, 3)
        17000
        
        >>> calculate_trip_cost(3, 4, 1)
        9800

    Assumptions:
        - The cost calculations are based on the given rates and do not account for any discounts or promotions.
        - The number of flights, nights, and guide days are independent of each other.
    """
    
    if flights < 0 or nights < 0 or guide_days < 0:
        raise ValueError("All inputs must be non-negative integers.")

    # Basic cost of the trip
    basic_cost = 500

    # Cost per flight
    flight_cost = flights * 1000

    # Cost per night stay
    night_cost = nights * 300

    # Daily cost for a tour guide
    guide_cost = guide_days * 200

    # Total trip cost is the sum of all costs
    total_trip_cost = basic_cost + flight_cost + night_cost + guide_cost

    return total_trip_cost


# Example usage:
print(calculate_trip_cost(2, 5, 3))  # Output: 17000
print(calculate_trip_cost(3, 4, 1))  # Output: 9800
```

# SORU5

Senaryo: Bir yazılım geliştirme projesinin maliyeti hesaplama

Kurallar:

Temel maliyet: 1000 TL (başlangıç ücreti).

Geliştirici maliyeti: her geliştirici için günlük 500 TL.

Proje süresi: her gün için toplam maliyet hesaplanır.

Her ek özellik için 200 TL ekler.

1. Soru: calculate_project_cost() fonksiyonunu yazın. Bu fonksiyon üç parametre alacak: developers (geliştirici sayısı), days (proje süresi 

gün olarak) ve features (ek özellik sayısı). Bu fonksiyon toplam maliyeti döndürecek.

        * 2 geliştirici, 10 gün ve 5 ek özellik için toplam maliyet nedir?

        * 1 geliştirici, 15 gün ve 10 ek özellik için toplam maliyet nedir?

        * 3 geliştirici, 7 gün ve 0 ek özellik için toplam maliyet nedir?

```python
def calculate_project_cost(developers: int, days: int, features: int) -> float:
    """
    Calculate the total cost of a software development project.

    Parameters:
    developers (int): The number of developers working on the project.
    days (int): The duration of the project in days.
    features (int): The number of additional features to be added to the project.

    Returns:
    float: The total cost of the project in TL. 

    Raises:
    ValueError: If either `developers` or `days` is less than zero.

    Assumptions:
    - Base cost of the project is 1000 TL.
    - Each developer costs 500 TL per day.
    - Each feature adds an additional 200 TL to the total cost.
    
    Example usage:
        >>> calculate_project_cost(2, 10, 5)
        34000.0
        >>> calculate_project_cost(1, 15, 10)
        67000.0
        >>> calculate_project_cost(3, 7, 0)
        9500.0
    """
    
    # Base cost of the project is 1000 TL.
    base_cost = 1000

    # Each developer costs 500 TL per day. If developers or days are negative,
    # raise a ValueError.
    if developers < 0:
        raise ValueError("Number of developers cannot be negative.")
    if days < 0:
        raise ValueError("Project duration cannot be negative.")

    developer_cost = developers * days * 500

    # Each feature adds an additional 200 TL to the total cost.
    features_cost = features * 200

    # Total project cost is the sum of base, developer and features costs
    total_cost = base_cost + developer_cost + features_cost

    return total_cost

# Testing the function with different scenarios

try:
    print(calculate_project_cost(2, 10, 5))
except Exception as e:
    print(str(e))

try:
    print(calculate_project_cost(1, 15, 10))
except Exception as e:
    print(str(e))

try:
    print(calculate_project_cost(3, 7, 0))
except Exception as e:
    print(str(e))
```
